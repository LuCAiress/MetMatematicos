```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Math
from math import pi
from scipy.integrate import quad
```

```python
def modulo_seno(t):
    return np.abs(np.sin(t))
```

```python
def serie_trig(t, n_harm):
    a0 = 2 / np.pi
    series = a0 / 2
    for n in range(1, n_harm + 1):
        an = (2 / np.pi) * ((-1)**n / (1 - 4*n**2))
        series += an * np.cos(2*n*t)
    return series
```

```python
display(Math(r'a_0 = \frac{2}{\pi}'))
display(Math(r'a_n = \frac{2}{\pi} \cdot \frac{(-1)^n}{1 - 4n^2}'))
display(Math(r'\left| \sin(t) \right| \approx \frac{a_0}{2} + \sum_{n=1}^{N} a_n \cos(2nt)'))
```

```python
#t = np.linspace(0, 2 * np.pi, 1000)
t = np.arange(-np.pi*2.5, pi*2.5, 0.001)
```

### Séries Trigonométricas

```python
plt.subplot(221)
plt.title(f'{3} harmônicos')
plt.plot(t, serie_trig(t, 3), color='red')
plt.xlabel('t')
plt.grid(True)

plt.subplot(222)
plt.title(f'{10} harmônicos')
plt.plot(t, serie_trig(t, 10), color='red')
plt.xlabel('t')
plt.grid(True)

plt.subplot(223)
plt.title(f'{50} harmônicos')
plt.plot(t, serie_trig(t, 50), color='red')
plt.xlabel('t')
plt.grid(True)

plt.subplot(224)
plt.title(f'{1000} harmônicos')
plt.plot(t, serie_trig(t, 1000), color='red')
plt.xlabel('t')
plt.grid(True)

plt.tight_layout()
plt.show()
```
