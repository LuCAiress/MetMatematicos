```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi
```

```python
#t = np.arange(-np.pi*2.5, pi*2.5, 0.001)
t = np.linspace(-np.pi*2, 2 * np.pi, 1000)
```

```python
def trig(t, nh):
    x = np.zeros(t.shape).astype('complex128')
    a0 = 2 / np.pi
    for n in range(1, nh, 1):
        an = (-4*np.cos(n*np.pi)**2)/((-1 + 4*n**2)*np.pi)
        x += an * np.cos(2 * n * t)
    return x + a0
```

```python
%%latex
$$a_0 = \frac{2}{\pi}$$
$$a_n = -\frac{4\cos(n\pi)^2}{(-1+4n^2)\pi}$$
$$|\sin(t)| \approx a_0 + \displaystyle\sum_{n=1}^{\infty} a_n\cos(2nt)$$
```

### Séries Trigonométricas

```python
def plot_trig(nh, t, sub):
    plt.subplot(sub)
    plt.title(f'{nh} harmônicos')
    plt.plot(t, trig(t, nh), color='red')
    plt.xlabel('t')
    plt.grid(True)

plot_trig(3, t, 221)
plot_trig(10, t, 222)
plot_trig(50, t, 223)
plot_trig(1000, t, 224)
plt.tight_layout()
```

### Séries Exponenciais

```python
def exp(t, nh):
    x = np.zeros(t.shape).astype('complex128')
    j = np.complex128(0 + 1j)
    for n in range(-nh, nh + 1):
        Dn = (-2*np.cos(n*np.pi)**2)/((-1 + 4*n**2)*np.pi)
        x += Dn * np.exp(j*2*n*t)
    return x
```

```python
%%latex
$$2*D_n=a_n-\cancel{(jb_n)}$$
$$D_n=\frac{a_n}2$$
$$D_n=-\frac{2\cos(n\pi)^2}{(-1+4n^2)\pi}$$
$$\displaystyle\sum_{n=-\infty}^{n=\infty}D_ne^{jn\omega_0t}
```

```python
def plot_exp(nh, t, sub):
    plt.subplot(sub)
    plt.title(f'{nh} harmônicos')
    plt.plot(t, exp(t, nh), color='red')
    plt.xlabel('t')
    plt.grid(True)

plot_exp(3, t, 221)
plot_exp(10, t, 222)
plot_exp(50, t, 223)
plot_exp(1000, t, 224)
plt.tight_layout()
```
