```python
from scipy.fft import fft, fftfreq, fftshift
import numpy as np
import matplotlib.pyplot as plt
```

### Função

```python
%%latex
$$e^{-|t|}\cdot\cos{10t}$$
```

### Transformada no domínio da frequência

```python
def transformada_freq(N):    
    t = np.linspace(-5, 5, N)
    y = np.exp(-np.abs(t)) * np.cos(10 * t)

    yf = fft(y)
    xf = fftfreq(N, (t[1] - t[0]))
    xf = fftshift(xf)
    yplot = fftshift(yf)

    plt.plot(xf, 1.0/N * np.abs(yplot))
    plt.title(f'{N} pontos')
    plt.xlabel("Frequência (Hz)")
    plt.ylabel("Amplitude") 
    plt.grid()
    plt.show()
```

```python
transformada_freq(256)
transformada_freq(512)
transformada_freq(1024)
transformada_freq(2048)
```

```python
def transformada_tempo(N):    
    t = np.linspace(-6, 6, N)
    y = np.exp(-np.abs(t)) * np.cos(10 * t)
    
    plt.plot(t, y)
    plt.title(f'{N} pontos')
    plt.xlabel("Tempo")
    plt.ylabel("Amplitude")
    plt.grid()
    plt.show()
```

```python
transformada_tempo(256)
transformada_tempo(512)
transformada_tempo(1024)
transformada_tempo(2048)
```
